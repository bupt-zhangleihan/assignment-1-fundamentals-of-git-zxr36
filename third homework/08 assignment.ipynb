```python
import pandas as pd
import matplotlib.pyplot as plt

#Loading the dataset
path = '../../data/form.csv'
form_df = pd.read_csv(path)

print(form_df['state'].dropna())

form_df['state'].value_counts().plot(kind='bar');
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Loading the dataset
path = '../../data/form.csv'
form_df = pd.read_csv(path)

form_df["birth_month"].replace({'JAN':'January','january':'January'}).value_counts().plot(kind='bar');
plt.show()
```
